```python
unique_counts = data.nunique()
print(unique_counts)
```

```python

y = data.Price
x = data.drop(['Price'], axis=1)
from sklearn.model_selection import train_test_split 
x_train, x_val, y_train, y_val = train_test_split(x,y,train_size = 0.8, test_size = 0.2, random_state = 0)

```

```python
from xgboost import XGBRegressor
my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, early_stopping_rounds=10)
my_model.fit(x_train, y_train, eval_set=[(x_val, y_val)], verbose=False)

from sklearn.metrics import mean_absolute_error,r2_score

predictions = my_model.predict(x_val)

print(mean_absolute_error(predictions, y_val))
```

```python

r2 = r2_score(y_val, predictions)
print("R-squared:", r2)
```

```python
import joblib

# Save the model
joblib.dump(my_model, 'house_price_model.pkl')

# Save the feature names
feature_names = x.columns.tolist()
with open('feature_names.txt', 'w') as f:
    for feature in feature_names:
        f.write(f"{feature}\n")
```
